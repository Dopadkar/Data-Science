### Проверка эффективности обогащения


Проверим, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака rougher.output.recovery. Найдите MAE между вашими расчётами и значением признака. Опишите выводы.

Формула рассчёта:

Для расчёта нам понадобятся:
    
- C - доля золота в концентрате после флотации/очистки;
- F - доля золота в сырье/концентрате до флотации/очистки;
- T - доля золота в отвальных хвостах после флотации/очистки.

Сохраним слльветствующие данне в переменных.

```python
train = data_[0]
# доля золота в концентрате после флотации/очистки
C_train = train['rougher.output.concentrate_au']
# доля золота в сырье/концентрате до флотации/очистки
F_train = train['rougher.input.feed_au']
# доля золота в отвальных хвостах после флотации/очистки
T_train = train['rougher.output.tail_au']
```

```python
# функция расчёта эффективности обогащения золота
def recovery(C, F, T):
    return 100 * (C * (F-T)) / (F * (C-T))
```

Произведём рассчёты.

```python
recovery_train = train['rougher.output.recovery']
# рассчитаем эффективность обогащения золота
recovery_calc = recovery(C_train, F_train, T_train)
```

Посмотрим на характеристики.

```python
recovery_train.describe()
```

```python
recovery_calc.describe()
```

Найдём среднюю абсолютную ошибку (MAE) между нашими расчётами и значением признака в таблице и процент MAE.

```python
mae = mean_absolute_error(recovery_train, recovery_calc)
mae_percentage = mean_absolute_percentage_error(recovery_train, recovery_calc)
print('MAE', mae)
print('MAE (%)', mae_percentage)
```

MAE близка к нулю.

**Выводы.** Зарактеристики рассчитанных и загруженных величин одинаковые, а MAE равна нудю, следовательно, эффективность обогащения золота указана точно.

#### Описание данных

**Технологический процесс**

- `Rougher feed` — исходное сырье
- `Rougher additions` (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
  
   - `Xanthate` — ксантогенат (промотер, или активатор флотации);
   - `Sulphate` — сульфат (на данном производстве сульфид натрия);
   - `Depressant` — депрессант (силикат натрия).
- `Rougher process` (англ. «грубый процесс») — флотация
- `Rougher tails` — отвальные хвосты
- `Float banks` — флотационная установка
- `Cleaner process` — очистка
- `Rougher Au` — черновой концентрат золота
- `Final Au` — финальный концентрат золота

**Параметры этапов**

- `air amount` — объём воздуха
- `fluid levels` — уровень жидкости
- `feed size` — размер гранул сырья
- `feed rate` — скорость подачи

**Наименование признаков**

Наименование признаков должно быть такое:

`[этап].[тип_параметра].[название_параметра]`

**Пример:** 

`rougher.input.feed_ag`

Возможные значения для блока [этап]:

- `rougher` — флотация
- `primary_cleaner` — первичная очистка
- `secondary_cleaner` — вторичная очистка
- `final` — финальные характеристики

Возможные значения для блока [тип_параметра]:

- `input` — параметры сырья
- `output` — параметры продукта
- `state` — параметры, характеризующие текущее состояние этапа
- `calculation` — расчётные характеристики

Шаблон: `[этап].[тип_параметра].[название_параметра]`

### Концентрация металлов (Au, Ag, Pb)

Посмотрим, как меняется концентрация металлов (Au, Ag, Pb) на различных этапах очистки.

```python
def concentrate(data, t):
    # обозначили две выборки конкретных столбцов с пользовательскими рейтингами.
    rougher = data['rougher.output.tail_' + t]
    primary_cleaner = data['primary_cleaner.output.concentrate_' + t]
    secondary_cleaner = data['secondary_cleaner.output.tail_' + t]
    final = data['final.output.concentrate_' + t]
    return rougher, primary_cleaner, secondary_cleaner, final
```

```python
def draw(rougher, primary_cleaner, secondary_cleaner, final):
    rougher.hist(bins=35, alpha=0.5, color='blue')
    primary_cleaner.hist(bins=35, alpha=0.5, color='red')
    secondary_cleaner.hist(bins=35, alpha=0.5, color='pink')
    final.hist(bins=35, alpha=0.5, color='green')
    plt.legend(['rougher', 'primary_cleaner', 'secondary_cleaner', 'final'])
    plt.show()
```

### Суммарная концентрация всех веществ

Исследуйте суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах. 

```python
def summary_rougher(row):
    try:
        rougher_ag = row['rougher.output.concentrate_ag']
    except:
        rougher_ag = 0
    try:
        rougher_pb = row['rougher.output.concentrate_pb']
    except:
        rougher_pb = 0
    try:
        rougher_sol = row['rougher.output.concentrate_sol']
    except:
        rougher_sol = 0
    try:
        rougher_au = row['rougher.output.concentrate_au']
    except:
        rougher_au = 0

    rougher = rougher_ag + rougher_pb + rougher_sol + rougher_au

    return rougher

def summary_rougher2(row):
    rougher_ag = row['rougher.input.feed_ag']
    rougher_pb = row['rougher.input.feed_pb']
    rougher_sol = row['rougher.input.feed_sol']
    rougher_au = row['rougher.input.feed_au']

    rougher = rougher_ag + rougher_pb + rougher_sol + rougher_au

    return rougher

def summary_final(row):
    try:
        final_ag = row['final.output.concentrate_ag']
    except:
        final_ag = 0
    try:
        final_pb = row['final.output.concentrate_pb']
    except:
        final_pb = 0
    try:
        final_sol = row['final.output.concentrate_sol']
    except:
        final_sol = 0
    try:    
        final_au = row['final.output.concentrate_au']
    except:
        final_au = 0

    final = final_ag + final_pb + final_sol + final_au

    return final
```

### Функция sMAPE

Напишем функцию для вычисления итоговой sMAPE.

Нужно спрогнозировать сразу две величины:
- эффективность обогащения чернового концентрата `rougher.output.recovery`;
- эффективность обогащения финального концентрата `final.output.recovery`.

```python
# Метрика sMAPE вычисляется так:
def calc_sMAPE(y, y_):
    try:
        N = y.shape[0]
        numerator = abs(y - y_)
        denominator = (abs(y) + abs(y_))
        part1 = (numerator / (denominator / 2))
        part2 = (part1 / N) 
        summary = (100 * part2).sum()
        return summary
    except:
        print(y, y_)
        print(y.shape[0], y.shape[1])
        print(y_.shape[0], y_.shape[1])

# Итоговая метрика складывается из двух величин
def final_sMAPE(rougher, rougher_, final, final_):
    return (0.25 * calc_sMAPE(rougher,rougher_) + 0.75 * calc_sMAPE(final, final_))
```

###### Кросс-валидация

```python
# функция для кросс-валидации
def cross_valid(features, target, model, n, save=False, scores=True):
    models = []
    scores = []
    j = 0
    # зададим размер блока, если их всего n
    sample_size = int(len(features)/n)
    for i in range(0, len(features)-1, sample_size):
        j+=1
        if j <= n:
            valid_indexes = list(range(i, i+sample_size))

            train_indexes = list(range(0, i)) + list(range(i+sample_size, len(features)))

            features_train = features.iloc[train_indexes]
            target_train = target.iloc[train_indexes]
            features_valid = features.iloc[valid_indexes]
            target_valid = target.iloc[valid_indexes]

            model = model.fit(features_train, target_train)
            predictions = model.predict(features_valid)
            sMAPE = calc_sMAPE(target_valid, predictions)

            scores.append(sMAPE)
            if save:
                models.append(model)
        else:
            break
        

    final_score = sum(scores) / len(scores)
    # print('Средняя оценка качества модели sMAPE:', final_score)
    return final_score
    if save:
        return models
```
